```python
import pandas as pd
```

```python
df=pd.read_csv('pandas_docs/survey_results_public.csv')
df
```

### Developers who gets more tham 18lacs, find their country, age and language worked with?


```python
a=df[df['ConvertedComp']>1800000].loc[:,['Country','LanguageWorkedWith']]
a
```

### If developer earns only less than 10000, find their respective country and their education background and working loc

```python
b=df[df['ConvertedComp']<10000].loc[:,['Country','EdLevel','WorkLoc']]
b
```

### - Developers who is not older than 18 years?

```python
df[df['ConvertedComp']>1500000]
```

### - Developers earns between 10000 and 12000?

```python
df[(df['ConvertedComp']>10000) & (df['ConvertedComp']<12000)]
```

### - Developers who earns more than 2000000 or earns 10000 find their WorkLoc and ConvertedComp?

```python
df[(df['ConvertedComp']>2000000) | (df['ConvertedComp']<10000)].loc[:,['WorkLoc','ConvertedComp']]
```

### - Developers who is older than 50 or more and gets 1500000 as remuneration ?

```python
df[(df['Age']>50) & (df['ConvertedComp']==1500000)]
```

### - Developers who earns more than 150000 or less than 10 years old ?

```python
df[(df['Age']<10) | (df['ConvertedComp']==150000)]
```

### - Developers who gets greater than or equal to 20 lakhs and 50 years old?

```python
df[(df['Age']==50) & (df['ConvertedComp']>=2000000)]
```


### - Developers who is not more than 20 years old or earns 1000 . if there someone find their Country,Age,WorkLoc and Hobbyist or not ?

```python
df[(df['Age']<=20) | (df['ConvertedComp']==1000)].loc[:,['Country','Age','WorkLoc','Hobbyist']]
```


### - Developers Who earns more than 1200000 or 15 years old find their ConvertedComp , CompTotal and age?

```python
df[(df['Age']==15) | (df['ConvertedComp']>1200000)].loc[:,['ConvertedComp','CompTotal','Age']]
```
